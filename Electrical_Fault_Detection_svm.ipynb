```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
import os
import zipfile
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.svm import SVC
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, precision_recall_curve, roc_curve, f1_score, precision_score, recall_score
from statsmodels.stats.outliers_influence import variance_inflation_factor
import kagglehub
```

```python
# Extract files
with zipfile.ZipFile("electrical-fault-detection-and-classification.zip", "r") as zip_ref:
    zip_ref.extractall("electrical_fault_data")
```

```python
# Load the dataset dynamically
extracted_files = os.listdir("electrical_fault_data")
df = pd.read_csv(f"electrical_fault_data/{extracted_files[0]}")
```

```python
# Clean column names
df.columns = df.columns.str.strip()
```

```python
# Display basic information
print("Dataset Shape:", df.shape)
print("\nDataset Info:")
print(df.info())
print("\nColumns in dataset:", df.columns.tolist())
print("\nFirst five rows:")
print(df.head())
print("\nSummary statistics:")
print(df.describe())
```

```python
# Checking unique values for categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    print(f"\nUnique values in {col}: {df[col].nunique()}")
```

```python
# Step 2: Handle Missing Values
print("\nMissing Values Before Imputation:")
print(df.isnull().sum())
```

```python
# Drop columns with excessive missing values
threshold = 0.4  # Drop columns with more than 40% missing values
df.dropna(thresh=int(threshold * len(df)), axis=1, inplace=True)
```

```python
# Checking skewness and imputing missing values accordingly
for col in df.select_dtypes(include=["float64", "int64"]).columns:
    if df[col].isnull().sum() > 0:
        skewness = df[col].skew()
        print(f"Skewness of {col}: {skewness}")
        imputer = SimpleImputer(strategy="median" if abs(skewness) > 0.5 else "mean")
        df[[col]] = imputer.fit_transform(df[[col]])
```

```python
print("\nMissing Values After Imputation:")
print(df.isnull().sum())
```

```python
# Step 3: Data Visualization
plt.figure(figsize=(12,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(12,6))
df.hist(bins=30, figsize=(12, 8), edgecolor='black')
plt.suptitle("Feature Distributions")
plt.show()
```

```python
# Step 4: Outlier Detection using Box Plot
plt.figure(figsize=(12,6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.title("Box Plot for Outlier Detection")
plt.show()
```

```python
# Handling Outliers using Winsorization
def cap_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])

for col in df.select_dtypes(include=["float64", "int64"]).columns:
    cap_outliers(col)
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.title("Box Plot After Outlier Handling")
plt.show()
```

```python
# Step 5: Feature Engineering
target_column = "Output (S)"  # Changed target column to 'Output (S)'
if target_column not in df.columns:
    raise ValueError(f"Target column '{target_column}' not found in dataset. Available columns: {df.columns}")


X = df.drop(target_column, axis=1)
y = df[target_column]
```

```python
# Apply SMOTE for class imbalance
smote = SMOTE(random_state=42)
X, y = smote.fit_resample(X, y)
```

```python
# Check for multicollinearity
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print("\nVariance Inflation Factor:")
print(vif_data)
```

```python
# Dimensionality Reduction using PCA
pca = PCA(n_components=0.95)  # Retain 95% variance
X_pca = pca.fit_transform(X)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_pca)
```

```python
# Step 6: Train-Test Split with Stratification
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, stratify=y, random_state=42)
```

```python
# Step 7: Hyperparameter Tuning for SVM
param_grid = {'C': [0.1, 1, 10, 100], 'gamma': ['scale', 'auto', 0.01, 0.1, 1], 'kernel': ['rbf']}
grid_search = GridSearchCV(SVC(probability=True, class_weight='balanced'), param_grid, cv=StratifiedKFold(n_splits=5), scoring='accuracy')
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print(f"\nBest Parameters: {best_params}")

svm_model = SVC(**best_params, probability=True, class_weight='balanced')
svm_model.fit(X_train, y_train)
```

```python
# Step 8: Model Evaluation
y_pred = svm_model.predict(X_test)
print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# Confusion Matrix Heatmap
plt.figure(figsize=(6,6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix Heatmap")
plt.show()
```

```python
# Save Model and Scaler
with open("svm_model.pkl", "wb") as file:
    pickle.dump((scaler, svm_model), file)

print("Model and scaler saved as svm_model.pkl")
```
